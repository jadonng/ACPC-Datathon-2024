```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.pipeline import Pipeline
```

```python
# meat prodution dataframe

monthly_df = pd.read_csv('Data/Meat_Stats_Meat_Production.csv')
monthly_df = monthly_df.sort_index(ascending=False)
monthly_df
```

```python
# jadon baseline df

baseline_monthly_df = pd.read_csv('Intermediate_data/Baseline_jadon_v1.csv')

baseline_monthly_df['Date-Time'] = pd.to_datetime(baseline_monthly_df['Date-Time'])
baseline_monthly_df['Year'] = baseline_monthly_df['Date-Time'].dt.year
baseline_monthly_df['Month'] = baseline_monthly_df['Date-Time'].dt.month
baseline_monthly_df['Date'] = baseline_monthly_df['Year'].astype(str) + "-" + baseline_monthly_df['Month'].astype(str)
baseline_monthly_df = baseline_monthly_df[["Date"] + [col for col in baseline_monthly_df.columns if col != "Date"]]
baseline_monthly_df = baseline_monthly_df.drop(['Date-Time', 'Year', 'Month'], axis=1)
baseline_monthly_df
```

```python
monthly_df['Animal'].unique()
```

```python
### meat dataframe prepration

type_of_meat = 'Poultry'
animal_list = ['Beef', 'Veal']

# meat_monthly_df = monthly_df[monthly_df['Type of Meat'] == type_of_meat]
meat_monthly_df = monthly_df[monthly_df['Animal'].isin(animal_list)]
meat_monthly_df['Production'] = meat_monthly_df['Production'].str.replace(',', '').astype(float)
meat_monthly_df = meat_monthly_df.groupby(['Date']).sum().reset_index()
meat_monthly_df = meat_monthly_df[['Date', 'Production']].reset_index().drop('index', axis=1)
meat_monthly_df['Date'] = pd.to_datetime(meat_monthly_df['Date'], format='%b-%Y').dt.strftime('%Y-%m')
meat_monthly_df = meat_monthly_df.sort_values('Date').reset_index().drop('index', axis=1)
meat_monthly_df
```

```python
other_monthly_df = monthly_df[monthly_df['Type of Meat'] != type_of_meat]
other_monthly_df['Production'] = other_monthly_df['Production'].str.replace(',', '').astype(float)
other_monthly_df = other_monthly_df.groupby(['Date']).sum().reset_index()
other_monthly_df = other_monthly_df[['Date', 'Production']].reset_index().drop('index', axis=1)
other_monthly_df['Date'] = pd.to_datetime(other_monthly_df['Date'], format='%b-%Y').dt.strftime('%Y-%m')
other_monthly_df = other_monthly_df.sort_values('Date').reset_index().drop('index', axis=1)
other_monthly_df
```

# Pipeline - Monthly Data (dataframe for prediction)

```python
class generate_class(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
        return self
    def transform(self, X):
        return X

class monthly_pct_change(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
        return self
    def transform(self, X):
        X['1m_pct_change_production'] = X['Production'].pct_change(1)
        X['3m_pct_change_production'] = X['Production'].pct_change(3)
        X['6m_pct_change_production'] = X['Production'].pct_change(6)
        return X
    
class monthly_vol(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
        return self
    def transform(self, X):
        X['3m_production_vol'] = X['Production'].rolling(window=3).std() * np.sqrt(12)
        X['6m_production_vol'] = X['Production'].rolling(window=6).std() * np.sqrt(12)
        return X
```

```python
df_cols = meat_monthly_df.columns

support_pipeline = Pipeline([
        # ('generate_class', generate_class()),
        ('monthly_pct_change', monthly_pct_change()),
        ('monthly_vol', monthly_vol())
        
])

baseline_df = support_pipeline.fit_transform(meat_monthly_df)
other_df = support_pipeline.fit_transform(other_monthly_df)
baseline_df = baseline_df.drop(['Production'],axis=1)
other_df = other_df.drop(['Production'],axis=1)
baseline_df
other_df
```

```python
tmp = pd.merge(baseline_df, baseline_monthly_df, on='Date')
tmp.to_csv('Intermediate_data/baseline+beef_veal_production_v1.csv', index=None)
```
